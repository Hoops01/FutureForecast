### Define custom functions

```python
def max_less_avg(arr):
    return arr.max() - arr.mean()
def per75(arr):
    return arr.quantile(0.75)
def per75_less_mean(arr):
    return arr.quantile(0.75) - arr.mean()
def percap(group):
    maxavg = group['avg']
    maxper = group['per75']
    maxcap = group['max']
    return min((maxper-maxavg).max(),maxcap.max()-maxavg.max())
```
